## Why Test Your Code

---

**Every developer has experienced this:** you fix a bug, deploy to production, and suddenly something else breaks. Without tests, you're flying blind.

<br>

### The Cost of Bugs

---

The later you find a bug, the more expensive it is to fix:

| Stage | Cost to Fix | Example |
|-------|-------------|----------|
| **During development** | Minutes | You notice the typo immediately |
| **In code review** | Hours | Reviewer spots the logic error |
| **In QA/staging** | Days | Testing reveals edge case failure |
| **In production** | Weeks+ | Customer reports data corruption |

Tests catch bugs at the **cheapest stage** - during development.

<br>

### Benefits of Testing

---

**1. Confidence to refactor**

Want to improve that messy function? With tests, you can refactor fearlessly - if tests pass, your changes are safe.

**2. Documentation through examples**

Tests show exactly how your code should be used. They're living documentation that can't get out of date.

**3. Regression prevention**

Once you fix a bug, write a test for it. That bug will never return unnoticed.

**4. Better design**

Code that's easy to test is usually well-designed. Testing encourages small, focused functions with clear responsibilities.

<br>

## Unit Tests vs Integration Tests

---

Not all tests are equal. Understanding the difference helps you write the right tests.

<br>

### Unit Tests

---

**Unit tests** test a single piece of code in isolation - typically one function or method.

**Characteristics:**
- Fast (milliseconds)
- Test one thing at a time
- No external dependencies (database, network, filesystem)
- Easy to pinpoint failures

````python
# Example: A function we want to test

def validate_email(email: str) -> bool:
    """Check if email format is valid."""
    if not email or '@' not in email:
        return False
    
    local, domain = email.rsplit('@', 1)
    
    if not local or not domain:
        return False
    
    if '.' not in domain:
        return False
    
    return True


# Try it in Python shell:
# >>> validate_email("user@example.com")
# True
# >>> validate_email("invalid")
# False
# >>> validate_email("user@")
# False
````

````python
# Unit test for validate_email

def test_validate_email_valid():
    assert validate_email("user@example.com") == True

def test_validate_email_missing_at():
    assert validate_email("userexample.com") == False

def test_validate_email_missing_domain():
    assert validate_email("user@") == False
````

<br>

### Integration Tests

---

**Integration tests** test how multiple components work together.

**Characteristics:**
- Slower (seconds to minutes)
- Test component cooperation
- May involve real database, API calls, filesystem
- Catch issues that unit tests miss

### The Testing Pyramid

---

The **testing pyramid** guides how many tests of each type to write:

```
        /\          E2E Tests (few)
       /  \         - Full user workflows
      /----\        - Slowest, most expensive
     /      \
    / Integr. \     Integration Tests (some)
   /   Tests   \    - Component cooperation
  /-------------\   - Medium speed/cost
 /               \
/   Unit Tests    \ Unit Tests (many)
-------------------  - Single functions
                     - Fast, cheap
```

**Write many unit tests**, some integration tests, and few end-to-end tests.

### pytest vs unittest

---

Python has a built-in `unittest` module, but pytest is preferred in modern Python:

| Aspect | unittest | pytest |
|--------|----------|--------|
| **Assertions** | `self.assertEqual(a, b)` | `assert a == b` |
| **Test class** | Required | Optional |
| **Setup/teardown** | `setUp`/`tearDown` methods | Fixtures |
| **Error messages** | Basic | Detailed with diffs |
| **Boilerplate** | More | Less |

````python
# unittest style - verbose
import unittest

class TestEmailValidation(unittest.TestCase):
    def test_valid_email(self):
        self.assertTrue(validate_email("user@example.com"))
    
    def test_invalid_email(self):
        self.assertFalse(validate_email("invalid"))
````

````python
# pytest style - simple

def test_valid_email():
    assert validate_email("user@example.com") == True

def test_invalid_email():
    assert validate_email("invalid") == False
````

`pytest` can also run unittest-style tests, so you can migrate gradually.

### Writing a Simple Test

---

**Step 1: Create your code** (`validators.py`):

````python
# validators.py

import re
from dataclasses import dataclass, field


@dataclass
class ValidationResult:
    is_valid: bool
    errors: list[str] = field(default_factory=list)


def is_valid_author(value: str) -> bool:
    return bool(re.match(r'^[a-zA-Z ]+$', value))


def has_no_spam_words(value: str) -> bool:
    spam_words = ('buy now', 'click here', 'free money')
    return not any(word in value.lower() for word in spam_words)


def validate_blog_post(title: str, author: str, content: str) -> ValidationResult:
    errors = []

    if not title:
        errors.append("Title is required")

    if not is_valid_author(author):
        errors.append("Author must contain only letters and spaces")

    if not has_no_spam_words(content):
        errors.append("Content looks like spam")

    return ValidationResult(is_valid=len(errors) == 0, errors=errors)
````

**Step 2: Create your test** (`tests.py`):

````python
# tests.py

from script import validate_blog_post


def test_valid_post():
    result = validate_blog_post(
        title="Django Testing Guide",
        author="John Doe",
        content="This is a great blog post about Django testing."
    )
    assert result.is_valid == True
    assert result.errors == []


def test_author_with_numbers():
    result = validate_blog_post(title="My Post", author="John123", content="Clean content.")
    assert result.is_valid == False
    assert "Author must contain only letters and spaces" in result.errors


def test_author_with_special_chars():
    result = validate_blog_post(title="My Post", author="John@Doe", content="Clean content.")
    assert result.is_valid == False
    assert "Author must contain only letters and spaces" in result.errors


def test_no_spam_valid():
    result = validate_blog_post(
        title="My Post",
        author="John Doe",
        content="This is a great blog post about Django testing."
    )
    assert result.is_valid == True


def test_spam_content():
    result = validate_blog_post(title="My Post", author="John Doe", content="buy now!")
    assert result.is_valid == False
    assert "Content looks like spam" in result.errors


def test_multiple_errors():
    result = validate_blog_post(title="", author="John123", content="buy now!")
    assert result.is_valid == False
    assert len(result.errors) == 3
````

### 🧠 EXERCISE 🧠, Write Tests for a Password Validator

---

You're given a password validator function. Write tests for it.

<br>

**The code to test** (`src/password.py`):

````python
# src/password.py

def is_strong_password(password: str) -> tuple[bool, list[str]]:
    """
    Check if password meets strength requirements.
    
    Requirements:
    - At least 8 characters
    - Contains at least one uppercase letter
    - Contains at least one lowercase letter
    - Contains at least one digit
    - Contains at least one special character (!@#$%^&*)
    
    Returns:
        Tuple of (is_valid, list_of_missing_requirements)
    """
    missing = []
    
    if len(password) < 8:
        missing.append("at least 8 characters")
    
    if not any(c.isupper() for c in password):
        missing.append("uppercase letter")
    
    if not any(c.islower() for c in password):
        missing.append("lowercase letter")
    
    if not any(c.isdigit() for c in password):
        missing.append("digit")
    
    special_chars = "!@#$%^&*"
    if not any(c in special_chars for c in password):
        missing.append("special character (!@#$%^&*)")
    
    return (len(missing) == 0, missing)
````
